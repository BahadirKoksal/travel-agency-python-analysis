
### Bir fonksiyon oluşturun


`get_key_value_from_city` fonksiyonunu oluşturun.


Fonksiyon iki argüman almalıdır:
- `city`
- `key`


Çıktı, şehir için karşılık gelen değer olmalıdır.


<br>


Örneğin, aşağıdakini çalıştırırsak:


`get_key_value_from_city(city=paris_city, key="total_population")`


Çıktı `2102650` olmalıdır.


<br>


<i><u>Not:</u> Bir anahtar mevcut değilse, işlev "_\<key\>_ tanımlanmamış" sonucunu döndürmelidir.

💡 **Fonksiyonumuzu kodlamaya başlamadan önce, onu daha küçük adımlara ayıralım.**


Buna ***sözde kodlama*** diyoruz ve bu, bir problemi koda çevirmeden önce onu parçalara ayırmak için yararlı bir yöntemdir.


1. Fonksiyonun adını ve alması gereken argümanları tanımlayın


1. Anahtar mevcut olmadığında bir sonuç alın


1. Anahtar mevcut olduğunda bir sonuç alın


1. Sonucu çıktı olarak verin


3. adımda, belirli bir şehir sözlüğünde belirli bir anahtarın varlığını kontrol eden bir `if ifadesi` gerekecektir.


Öncelikle, yukarıda verilen test örneğini kullanarak if ifadesini nasıl yazacağımızı hatırlayalım!


`paris_city` sözlüğünde `total_population` anahtarını kontrol eden ve (varsa) bu anahtarın değerini yazdırılan bir if ifadesi yazın.


Sonuç `2102650` olmalıdır.

```python
if "total_population" in paris_city.keys():
    print(paris_city["total_population"])
```


Harika! Artık belirli bir sözlükteki belirli bir anahtara nasıl erişeceğimizi bildiğimize göre, bir fonksiyon kullanarak bunu *herhangi bir* sözlükteki *herhangi bir* anahtar için çalıştıracağız. Fonksiyonları bu kadar kullanışlı kılan da budur! Fonksiyonlar, aynı kodun tek bir amaç için değil, tekrar tekrar kullanılmasını sağlar.


İşlevimiz için, 3. adımda yazdığımız `if ifadesi`ne ihtiyacımız olacak, ancak elbette çıktının her zaman `paris_city[total_population]` olmasını istemeyiz!


`paris_city` kelimesini kullandığınız yerlerde bunu `city` ile, `total_population` kelimesini kullandığınız yerlerde ise `key` ile değiştirmemiz gerekecek.


İşlev daha sonra bu argümanları değişken olarak kullanacak. Hatta her bir argümanı işlevimiz içinde istediğimiz kadar kullanabiliriz! Bunu aşağıdaki resimde gösterdik 🤯


![Bir argümanın fonksiyon içinde nasıl kullanıldığına dair örnek](https://drive.google.com/uc?export=view&id=11ZttqXVheSMeoM0hI6zNHBjc5Bl5Uqb6&w=1000)


Fonksiyonlarımızda argümanları bu şekilde kullanıyoruz 🤯

```python

def get_key_value_from_city(city, key):
    result = f"{key} tanımlanmamış"
    if key in city:
        result = city[key]
    return result

print(get_key_value_from_city(city=paris_city, key="total_population"))
```

### Okuma değeri



Önceki `get_key_value_from_city` işlevini kullanarak, iki argüman alan yeni bir işlev oluşturun: `city_name` ve `key`.


Şehir listede yoksa, işlev "*şehir* mevcut değil veya kapsanmıyor" sonucunu döndürmelidir.


İşlevinizi birden fazla örnekle test edin.

```python
def retrieve_city_key_value_database(city_name, key):
    # varsayılan çıktı
    result = "şehir mevcut değil veya kapsanmıyor"

    # şehirler üzerinde döngü başlat
    for city in cities:
        # seçilen şehirde gerekli anahtar varsa çıktıyı ayarla
        if city["name"] == city_name:
            result = get_key_value_from_city(city, key)

    # test
    return result

print(retrieve_city_key_value_database("Paris", "total_population"))
print(retrieve_city_key_value_database("Tokyo", "total_population"))
```

### Değeri güncelle


Bazen şehir bilgileri yanlış olabilir veya güncellenmesi gerekebilir.


Mevcut bir anahtarın değerini yenisiyle değiştiren bir `modify_city_key_value` işlevi oluşturun. \
Bu işlev üç argüman alır: bir `city_name`, bir `key` ve anahtara atanacak yeni bir `value`.


İşlev, aşağıdaki senaryolardan birini belirten bir metin mesajı oluşturacaktır:


- Şehir mevcut değilse veya kapsanmıyorsa: "*şehir* mevcut değil veya kapsanmıyor."
- Sağlanan anahtar mevcut değilse: "*anahtar* mevcut değil."
- Anahtar şehir için başarıyla güncellendiyse: "*anahtar* *şehir* için güncellendi."

```python
def modify_city_key_value(city_name, key, value):
    for city in cities:
        if city["name"] == city_name:
            if key in city:
                city[key] = value
                return f"{key} {city_name} için güncellendi."
            else:
                return f"{key} mevcut değil."
    return f"{city_name} mevcut değil veya kapsanmıyor."

# test
print(modify_city_key_value("Paris", "total_population", 2200000))
print(modify_city_key_value("Paris", "mayor", "Anne Hidalgo"))
print(modify_city_key_value("Tokyo", "total_population", 1000))
```

### Arama motorunu tasarlayın


İki argüman alan `get_cities_nearby` işlevini oluşturun:
- `customer_location`: [enlem, boylam] biçiminde bir koordinat listesi




- `radius`: km cinsinden eşik değeri


Ve sağlanan customer_location yakınında bulunan `şehirlerin listesini` döndürür.


Hiçbir şehir bulunamazsa, işlev boş bir liste döndürmelidir.

İşlevinizi test etmek için Çin'in Guangyan şehrini (koordinatlar [32.4355, 105.8436]) ve 500 km'lik bir yarıçapı kullanabilirsiniz.


Xi'an ve Chengdu'nun çok uzak olmadığını göreceksiniz...

```python
def get_cities_nearby(customer_location, radius):
    nearby_cities = []

    for city in cities:
        city_location = city["location"]
        dist = distance.distance(customer_location, city_location).km

        if dist <= radius:
            nearby_cities.append(city["name"])

    return nearby_cities
```

Tebrikler! 🎉 \
Artık seyahat acentelerinin kullanabileceği bir arama fonksiyonuna sahipsiniz!

```python
print(get_cities_nearby([32.4355, 105.8436], 500))
```
